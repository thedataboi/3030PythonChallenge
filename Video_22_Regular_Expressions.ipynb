# Regex...is a game changer
I meant that, getting you head around Regex will change the game...but all game changers it's hard in partice and you'll need to use it often often and repeatedly if you want it to stick in your head.

https://pythex.org/ helps you validate your regex so you're not flying blind it also has a cheat sheet.

```python
#Create a pattern object containing the pattern you want to split
pattern = '[ \t\n\r\f\v]'
```

```python
#Store the pattern in a regex object
regex = re.compile(pattern)
```

```python
#split the text by the a blank space
play_data.description.apply(lambda x: regex.split(x)).head()
```

```python
#Create pattern to look for digits
pattern = '\d'
regex = re.compile(pattern)
```

```python
#Find all numberical data in the each body of text
play_data.description.apply(lambda x: regex.findall(x)).head()
```
