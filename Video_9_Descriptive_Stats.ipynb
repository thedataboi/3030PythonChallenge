## Summery Descriptive Stats<br>
Pandas have most basic calcuations as methods...follow on and see.

```python
#Sum for each column
play_data[['two_stars','one_star']].sum()
```

```python
#Get the the mean for each column
play_data[['two_stars','one_star']].mean(skipna=True)
```

```python
#Get the the mean across the columns
play_data[['two_stars','one_star']].mean(axis=1, skipna=True)
```

```python
#Gives the row label for the lowest value in the series
play_data['one_star'].idxmin()
```

```python
#Gives the row label for the highest value in the series
play_data['one_star'].idxmax()
```

```python
#Gives a series of the cummulative values of the dataframe
play_data['one_star'].cumsum()
```
