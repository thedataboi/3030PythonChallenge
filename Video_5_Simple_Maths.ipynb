# Maths, a quick guide to doing Maths in Pandas
If you're a marketer and not doing some for of calculation daily...I'm really confused to how you still have a job! We're going to cover the basic math operations to get you going with your KPI's.

```python
import pandas as pd #load pandas
```

```python
#Load dataset
play_data = pd.read_json('data/databoi_data_Set.json')
```

```python
#inspect the types of data you have
play_data.info()
```

```python
#Statistical description of data
play_data.describe().round(2)
```

```python
#Eyeball the data
play_data.head()
```

```python
#adding two columns in the order they're displayed
play_data['five_stars'].add(play_data['four_stars'])
```

```python
#adding the right column to the left
play_data['five_stars'].radd(play_data['four_stars'])
```

```python
#Subtract two columns in the order they're displayed
play_data['five_stars'].sub(play_data['four_stars'])
```

```python
#Subtract the right column to the left
play_data['five_stars'].rsub(play_data['four_stars'])
```

```python
#Divide two columns in the order they're displayed
play_data['five_stars'].div(play_data['four_stars'])
```

```python
##Floor Divide (return an integer not a float) two columns in the order they're displayed
play_data['five_stars'].floordiv(play_data['four_stars'])
```

```python
#Floor Divide (return an integer not a float) the right column to the left
play_data['five_stars'].rfloordiv(play_data['four_stars'])
```

```python
#Multiply two columns in the order they're displayed
play_data['five_stars'].mul(play_data['four_stars'])
```

```python
#Multiply the right column to the left
play_data['five_stars'].rmul(play_data['four_stars'])
```

```python
#The exponentiation of two columns in the order they're displayed
play_data['five_stars'].pow(play_data['four_stars'])
```

```python
#The exponentiation of the right column to the left
play_data['five_stars'].rpow(play_data['four_stars'])
```
