# Looking Up, matching and munging string data
If you're dealing with a lot of string data you'll need to know how to quickly look up text within a dataset. Pandas has some handy and scalable solutions for dealing with the tricky problem of text.

```python
#Lets find out if any of our data is null
play_data.description.isnull() #.nunique()
```

```python
play_data.description.str.contains('best')
```

```python
#Create pattern to look for digits
pattern = '\d'
regex = re.compile(pattern)
```

```python
play_data.description.str.findall(pattern)
```

```python
#Create pattern to look for text between quote marks
pattern = '\"(.+?)\"'
regex = re.compile(pattern)
```

```python
#Find all numberical data in the each body of text
play_data[play_data.description.str.match(pattern)]
```
