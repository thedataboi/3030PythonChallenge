# Random Sampling<br>
Random sampling that is useful for running test or if you want to see the dataset in a way which will reduce the change of repeatedly viewing the same data that you did previous...and Python makes this easy!

```python
#Create an object to select 20000 random samples
sampler = np.random.permutation(20000)
```

```python
sampler
```

```python
#Apply the random sampler to your dataframe using take
play_data.take(sampler)
```

```python
#The random sampler can also be applied this way. This method allows you to reduce the size of the random sample
play_data.sample(n=200)
```

```python
#In this variation you can create a object of your random sample 
new_sample = play_data.sample(n=200, replace=True)
```
