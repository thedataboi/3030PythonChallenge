## apply v's mapapply<br>
I'm going to quickly cover what apply does as well as the different situations where you'd use applymap instead of apply.

```python
#create a function using lambda which multiplies elements by 2
bit_of_math = lambda x: x*2
```

```python
play_data[['rating','rating_count']].head()
```

```python
#use .apply to perform the function on the selected series within the dataframe
play_data[['rating','rating_count']].apply(bit_of_math)
```

```python
#create a function using lambda which counts the length of a string
count_string = lambda x: len(str(x))
```

```python
#use .apply to perform the function on the selected series within the dataframe
play_data.select_dtypes('object').apply(count_string)
```

```python
#use .mapapply to perform the function on the series within the dataframe
play_data.select_dtypes('object').applymap(count_string)
```
