# Sorting and Ranking
Sorting is something we do all the time, it's not exactly a click of a button in Pandas like it is in Excel but it's pretty close and much faster on large datasets.

```python
import pandas as pd #load pandas
```

```python
#Load dataset
play_data = pd.read_json('data/databoi_data_Set.json')
```

```python
#inspect the types of data you have
play_data.info()
```

## Sorting & Ranking Data<br>
You will always want to sort your data and in some cases rank your data, here's how you'd do it in Python. We'll start with sorting.

```python
play_data.set_index('link', inplace=True)
```

```python
play_data.head()
```

```python
play_data.sort_index()
```

```python
play_data.sort_index(axis=1, ascending=False)
```

```python
play_data.two_stars.sort_values(ascending=False)
```

```python
play_data.sort_values(by='two_stars')
```

```python
play_data.sort_values(by=['two_stars','one_star'])
```

### Ranking Data<br>
You will always want to sort your data and in some cases rank your data, here's how you'd do it in Python.

```python
play_data['new_rank'] = play_data.rating.rank(ascending=1)
```

```python
play_data.sort_values(by='new_rank')
```
