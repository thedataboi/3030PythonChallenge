# Basic Filtering in Pandas
Filtering is a requirement on a daily basis. Here's how to filter in Pandas, you see it's super quick on large datasets.

```python
import pandas as pd #load pandas
```

```python
#Load dataset
play_data = pd.read_json('data/databoi_data_Set.json')
```

```python
#inspect the types of data you have
play_data.info()
```

```python
#Statistical description of data
play_data.describe().round(2)
```

```python
#Eyeball the data
play_data.head()
```

```python
#Filter App name and Developer name by a rating above 4
play_data[['developer_name', 'app_name']][play_data['rating'] >= 4]
```

```python
#Show me if a App is in the Tools category or not
play_data.category != 'Tools'
```

```python
#Display the above as a dataframe
play_data[play_data.category != 'Tools']
```

```python
#Show me a dataframe with category containing Health & Fitness but only apps with rating above 4
play_data[(play_data.category == 'Health & Fitness') &  (play_data['rating'] >= 4)]
```
