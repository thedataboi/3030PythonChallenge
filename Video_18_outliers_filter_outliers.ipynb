## Detecting outliers<br>
Everyone now and again you'll get a data point that lives out the normal range of an activity this is an outlier. For example you're usually order values is £50 but one day someone spent £200. You might want to filter this outlier because it's not representitive of a normal day and could skew any decision making.

```python
#Create an object which is only the column we want to exclude the outliers from
num_ratings = play_data.rating_count
num_ratings
```

```python
num_ratings.mean()
```

```python
# Filter the series for any values less 15000
num_ratings[num_ratings < 15000]
```

```python
#Create a dataframe of only numerical data
numeric_data = play_data.select_dtypes(['float','int'])
numeric_data.head()
```

```python
#Return a dataframe where all the values are below 15000
numeric_data[(numeric_data < 15000).all(1)]
```
