# Grouping your data
GroupBy is incredibly useful for understanding the ranges of your data cut by a specific segment.

```python
#Group the data by the price and look at the mean
play_data.groupby(by='price').mean()
```

```python
#Create an object which is a groupby of category and it's mean and round it to 2
category_mean = play_data.groupby(by='category').mean().round(2)
```

```python
category_mean.head()
```

```python
#Use stack to change the layout of thedata
category_mean.stack()
```

```python
#Use unstack to change the layout of the data to the inverse of stack
category_mean.unstack()
```

```python
#Plot a scatter of the data which has been grouped
category_mean.plot.scatter('five_stars','rating_count',figsize=(20,10))
```

```python
#Plot a bar chart of average rating count by category
category_mean.rating_count.plot.bar(figsize=(20,10))
```
