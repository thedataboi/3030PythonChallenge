## Replacing Values?<br>
I can't think of a month going by where I've not had to replace a value...even it it's with something as simple as a space..or even removing a space, it's all replacing

```python
#replace all the NaN values with a single number
play_data.replace('NaN', 0) #,inplace=True to replace the view
```

```python
#replace values with a dictionary
play_data['age_restriction'].replace({0:'No',1:'Yes'}) #,inplace=True to replace the view
```

```python
#replace values using list
play_data['age_restriction'].replace([0,1],['No','Yes']) #,inplace=True to replace the view
```
