## Selecting Series of data

```python
#Select a single series
play_data['developer_name']
```

```python
#Select more than one series
play_data[['developer_name', 'rating']]
```

Loc Vs iLoc<br>
Loc = is for label based Index's
iLoc = is for Integer based Index's

```python
#Returns data with the label 1000 for the columns 'developer_name' and 'rating'
play_data.loc[1000, ['developer_name', 'rating']] #loc[row label,[column name]]
```

```python
#Returns data up to and including label 1000 for the columns 'developer_name' and 'rating'
play_data.loc[:1000, ['developer_name', 'rating']] #loc[row label range,[column name]]
```

```python
#Display the data in all the series contained in row 1000
play_data.iloc[1000] #iloc[row position]
```

```python
#Returns data in the row 1000 for the columns 'developer_name' and 'rating'
play_data.iloc[[1000], [7, 30]] #iloc[[Row postion],[Column postion]]
```

```python
#Returns data in rows 1000 and 320000 for the columns 'developer_name' and 'rating'
play_data.iloc[[1000,32000], [7, 30]] #iloc[[Row postion],[Column postion]]
```

```python
play_data.iloc[10:1000, 4:][play_data.rating > 3] #iloc[rows, series][filter]
```

```python
# Select rows between 300000 and the final row
play_data[300000:-1]
```
