# Quick Plots in Pandas
Matplotlib is the daddy package but if all you need to do is quickly see the split or a direction of a line the Pandas has a built in package to help do this.

```python
#plot a cumluative graph of the first 200 rows of data for Trust
maj_data.trust[:200].cumsum().plot()
```

```python
#Plot the entire dataframe cummulative values
maj_data[:200].cumsum().plot()
```

```python
#Plot with a secondary axis
maj_data[:200].cumsum().plot(secondary_y='vol')
```

```python
#Plot bar values of Cit values
maj_data.cit[:100].plot.bar(figsize=(20,10))
```

```python
#Plot horizontal bar values of Cit values
maj_data.cit[:100].plot.barh(figsize=(20,15))
```

```python
#Plot horizontal bar values of Cit values, colour the chart black
maj_data.cit[:100].plot.barh(figsize=(20,15), color='k')
```

```python
#set an index for the maj_data dataset
maj_data.set_index('maj_link', inplace=True)
```

```python
#Name the index of the dataset
maj_data.index.name = 'App URL'
```

```python
#Draw a bar chart of the first 10 apps, with a secondary Y
maj_data[:10].plot.bar(secondary_y='vol')
```

```python
#Draw a horizontal stacked bar chart of the first 10 apps, with a secondary Y 
maj_data[:10].plot.barh(secondary_y='vol', stacked=True)
```

```python
#Draw a horizontal, transparent, stacked bar chart of the first 10 apps, with a secondary Y 
maj_data[:10].plot.barh(secondary_y='vol', stacked=True, alpha=0.5)
```
