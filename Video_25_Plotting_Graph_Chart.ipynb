# Hello...Matplotlib
Matplotlib is the go to drawing package, there are other packages but it's best to start with Matplotlib as some of the other packages you'll use are built on top of this. 

```python
#merge the link data from majestic to the play store data by creating a new object
fig = plt.figure(figsize=(20,5))
ax1 = fig.add_subplot(2,2,1)
ax2 = fig.add_subplot(2,2,2)

ax1.plot(maj_data.vol)
ax2.plot(maj_data.cit)
```

```python
#merge the link data from majestic to the play store data by creating a new object
fig = plt.figure(figsize=(20,5))
ax1 = fig.add_subplot(2,2,1)
ax2 = fig.add_subplot(2,2,2)

ax1.plot(maj_data.vol, color='magenta', linestyle='--')
ax2.plot(maj_data.cit, color='cyan')
```

```python
#Drawing a single graph and changing the format
fig = plt.figure(figsize=(20,5))
ax = fig.add_subplot(1,1,1)


ax.plot(maj_data['trust'][:30], 'rx--')
```

```python
#Another way to change the format
fig = plt.figure(figsize=(20,5))
ax = fig.add_subplot(1,1,1)

ax.plot(maj_data['trust'][:30], color='red', marker='x', linestyle='--')
```

```python
#More amendments to the style of the graph
fig = plt.figure(figsize=(20,5))
ax = fig.add_subplot(1,1,1)


ax.plot(maj_data['trust'][:30], color='red', marker='x', linestyle='--', label='Default', drawstyle='steps-post')
ax.legend(loc='best')
```

```python
#Give the graph new x labels, x scale and  and a title 
fig = plt.figure(figsize=(20,5)) #create the graph size
ax = fig.add_subplot(1,1,1) # Create a single plot space


ax.plot(maj_data['trust'][:30], color='red', marker='x', linestyle='--', label='Default', drawstyle='steps-post')
ax.legend(loc='best')


tick = ax.set_xticks([0, 10, 20, 30, 40]) # the spacing between labels
labels = ax.set_xticklabels(['Zero', 'Ten', 'Twenty', 'Thirty', 'Forty'], rotation=30, fontsize='small') # the label names
ax.set_title('Majestic Data Trust Flow first 30 Rows')
ax.set_xlabel('Apps')
```

```python
#Add Annotation to your graph
fig = plt.figure(figsize=(20,5)) #create the graph size
ax = fig.add_subplot(1,1,1) # Create a single plot space


ax.plot(maj_data['trust'][:30], color='red', marker='x', linestyle='--', label='Default', drawstyle='steps-post')
ax.legend(loc='best')


tick = ax.set_xticks([0, 10, 20, 30, 40]) # the spacing between labels
labels = ax.set_xticklabels(['Zero', 'Ten', 'Twenty', 'Thirty', 'Forty'], rotation=30, fontsize='small') # the label names
ax.set_title('Majestic Data Trust Flow first 30 Rows')
ax.set_xlabel('Apps')

ax.annotate('The App has a lot of Trust', xy=(8, 38), xytext=(10, 28),
            arrowprops=dict(facecolor='cyan', shrink=0.03, lw=2, ec='magenta')
           )
```

```python
#Save your graph as a png file
fig = plt.figure(figsize=(20,5)) #create the graph size
ax = fig.add_subplot(1,1,1) # Create a single plot space


ax.plot(maj_data['trust'][:30], color='red', marker='x', linestyle='--', label='Default', drawstyle='steps-post')
ax.legend(loc='best')


tick = ax.set_xticks([0, 10, 20, 30, 40]) # the spacing between labels
labels = ax.set_xticklabels(['Zero', 'Ten', 'Twenty', 'Thirty', 'Forty'], rotation=30, fontsize='small') # the label names
ax.set_title('Majestic Data Trust Flow first 30 Rows')
ax.set_xlabel('Apps')

ax.annotate('The App has a lot of Trust', xy=(8, 38), xytext=(10, 28),
            arrowprops=dict(facecolor='cyan', shrink=0.03, lw=2, ec='magenta')
           )

plt.savefig('file_name.png')
```
